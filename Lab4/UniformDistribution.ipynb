```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate a population (non-normal distribution: Uniform)
population = np.random.uniform(low=-5, high=5, size=100000)

# Plot population distribution
plt.hist(population, bins=100, density=True, color="green", alpha=0.7)
plt.title("Population Distribution (Uniform)")
plt.show()

# 2. Central Limit Theorem demonstration
sample_means = []
n = 30  # sample size
num_samples = 2000  # number of samples

for _ in range(num_samples):
    sample = np.random.choice(population, size=n, replace=True)
    sample_means.append(np.mean(sample))

# Plot sample means distribution
plt.hist(sample_means, bins=50, density=True, color="purple", alpha=0.7)
plt.title("Distribution of Sample Means (n=30)")
plt.show()
```
