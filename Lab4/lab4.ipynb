 # 1 Experiment the Central Limit Theorem

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# 1. Generate a population (Exponential distribution)
population = np.random.exponential(scale=2.0, size=100000)
```

```python
# Plot population distribution
plt.hist(population, bins=100, density=True, color="blue", alpha=0.7)
plt.title("Population Distribution (Exponential)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
# 2. Central Limit Theorem demonstration
sample_means = []
n = 50        # sample size
num_samples = 3000  # number of samples

for _ in range(num_samples):
    sample = np.random.choice(population, size=n, replace=True)
    sample_means.append(np.mean(sample))
```

```python

# 3. Plot sample means distribution
plt.hist(sample_means, bins=50, density=True, color="blue", alpha=0.6)

# Plot sample means distribution
plt.hist(sample_means, bins=50, density=True, color="blue", alpha=0.7)
plt.title("Distribution of Sample Means (n=50)")
plt.show()
```

# 2 Experiment the Law of Large Numbers

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate a population (Exponential distribution)
population = np.random.exponential(scale=2.0, size=1_000_000)

# True population mean
true_mean = np.mean(population)
print("True population mean:", true_mean)

# 2. Experiment with LLN
sample_sizes = np.arange(10, 50000, 500)
sample_means = []

for n in sample_sizes:
    sample = np.random.choice(population, size=n, replace=False)
    sample_means.append(np.mean(sample))

# 3. Visualization
plt.figure(figsize=(10,6))
plt.plot(sample_sizes, sample_means, label="Sample mean")
plt.axhline(true_mean, color="red", linestyle="--", label="True mean")
plt.xlabel("Sample size")
plt.ylabel("Mean")
plt.title("Law of Large Numbers (Exponential distribution)")
plt.legend()
plt.show()
```
