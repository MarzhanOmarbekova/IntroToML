```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate a population (non-normal distribution: Exponential)
population = np.random.exponential(scale=2.0, size=100000)

# Plot population distribution
plt.hist(population, bins=100, density=True, color="orange", alpha=0.7)
plt.title("Population Distribution (Exponential)")
plt.show()

# 2. Central Limit Theorem demonstration
sample_means = []
n = 50  # sample size
num_samples = 1000  # number of samples

for _ in range(num_samples):
    sample = np.random.choice(population, size=n, replace=True)
    sample_means.append(np.mean(sample))

# Plot sample means distribution
plt.hist(sample_means, bins=50, density=True, color="blue", alpha=0.7)
plt.title("Distribution of Sample Means (n=50)")
plt.show()
```
