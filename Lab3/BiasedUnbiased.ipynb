```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# Unbiased
axs[0].scatter([1.0, 1.1, 0.9, 1.05], [0.5, 0.6, 0.4, 0.55], color="blue")
axs[0].axvline(1, color="red", linestyle="--", label="True θ")
axs[0].set_title("Unbiased Estimator")
axs[0].legend()

# Biased
axs[1].scatter([1.3, 1.4, 1.2, 1.35], [0.5, 0.6, 0.4, 0.55], color="blue")
axs[1].axvline(1, color="red", linestyle="--", label="True θ")
axs[1].set_title("Biased Estimator")
axs[1].legend()

plt.show()
```
